```
#importando as bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
x=[1,2,4,5,6,8,9,10,11,12,14,15,20,25,30]
y=[3,5,16,29,38,61,83,130,125,150,169,220,380,730,850]
```

```
 df=pd.DataFrame(list(zip(x, y)), 
               columns =['X', 'Y'])
```

```
df.head()
```

```
plt.figure(figsize=(15,10))
plt.scatter(df.X,df.Y,c="b")
```

```
linear = np.poly1d(np.polyfit(df['X'], df.Y, 1)) #me retorna uma reta
```

```
quadratica = np.poly1d(np.polyfit(df.X, df.Y, 2)) # me retorna um polinomio de segundo grau que melhor explica os pontos
```

```
alto_grau = np.poly1d(np.polyfit(df.X, df.Y, 10)) # me retorna um polinomio de x graus
```

```
t = np.linspace(0, 30, 200) # me gera 200 valores
```

```
plt.figure(figsize=(15,10))
plt.plot(df.X, df.Y, 'o', t, linear(t), '-')
```

```
plt.figure(figsize=(15,10))
plt.plot(df.X, df.Y, 'o', t, quadratica(t), '-')
```

```
plt.figure(figsize=(15,10)) # vai gerar um exemplo de over
plt.plot(df.X, df.Y, 'o', t, alto_grau(t), '-')
```

```
from sklearn.metrics import r2_score #método para o cálculo do R2 (coeficiente de determinação)
R_2 = r2_score(df.Y, alto_grau(df.X))  #realiza o cálculo do R2

print("Coeficiente de Determinação (R2):", R_2)
```

```
previsao_segundo_grau=quadratica(27)
previsao_alto_grau=alto_grau(27)
```

```
print("Previsão quadrática: {}".format(previsao_segundo_grau))
print("Previsão décimo grau: {}".format(previsao_alto_grau))
```
